### ELECTRIC VEHICLE MARKET ANALYSIS

```python
#1. IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional Settings
plt.style.use('dark_background')
sns.set_style("darkgrid")
```

```python
# 2. LOAD DATASET

df = pd.read_csv("ev_market_2026.csv")
```

```python
# 3. BASIC DATA EXPLORATION

print("First 5 Rows")
print(df.head(5))
```

```python
print("Dataset Shape")
print(df.shape)
```

```python
print("Column Names")
print(df.columns)
```

```python
print("Statistical Summary")
print(df.describe())
```

```python
print("Dataset Information")
print(df.info())
```

```python
# 4. CHECK MISSING VALUES

print("Missing Values")
print(df.isnull().sum())
```

```python
# 5. REMOVE DUPLICATES

print("Duplicate Rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

```python
# 6. CHECK DATA TYPES

print("\nData Types")
print(df.dtypes)
```

```python
#7. Rename column 

df.rename(columns={'annual_sales_units':'Sales Units'}, inplace=True)
df.rename(columns={'price_usd':'Price'}, inplace=True)
df.rename(columns={'country_of_origin':'Country'}, inplace=True)
df.rename(columns={'battery_capacity_kwh':'Battery Capacity'}, inplace=True)
df.rename(columns={'charging_speed_kw':'Charging Speed'}, inplace=True)

print("Updated Columns")
print(df.columns)
```

```python
# 8. SAVE CLEANED DATASET

df.to_csv("clean_ev_market.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```

 ### EXPLORATORY DATA ANALYSIS (EDA)

```python
# 9. TOP BRANDS BY SALES

brand_sales = df.groupby('brand')['Sales Units'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
brand_sales.plot(kind='bar', color='lime')

plt.title("Top EV brands by Sales", fontsize=16)
plt.xlabel("brand")
plt.ylabel("Sales Units")

plt.xticks(rotation=45)

plt.show()
```

```python
# 10. PRICE DISTRIBUTION

plt.figure(figsize=(10,5))

sns.histplot(df['Price'], bins=20, kde=True, color='cyan')

plt.title("EV Price Distribution", fontsize=16)

plt.show()
```

```python
# 11. COUNTRY-WISE SALES

country_sales = df.groupby('Country')['Sales Units'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))

country_sales.plot(kind='bar', color='orange')

plt.title("Country-wise EV Sales", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Sales Units")

plt.xticks(rotation=45)

plt.show()
```

```python
# 12. MARKET SEGMENT SHARE

segment_sales = df.groupby('market_segment')['Sales Units'].sum()

plt.figure(figsize=(8,8))

plt.pie(
    segment_sales,
    labels=segment_sales.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Market Segment Share")

plt.show()
```

```python
# 13. BATTERY vs RANGE

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Battery Capacity',
    y='range_miles',
    hue='brand',
    data=df
)

plt.title("Battery Capacity vs Driving Range")

plt.show()
```

```python
# 14. PRICE vs CUSTOMER RATING

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Price',
    y='customer_rating',
    hue='market_segment',
    data=df
)

plt.title("Price vs Customer Rating")

plt.show()
```

```python
# 15. TOP RATED BRANDS

top_rated = df.groupby('brand')['customer_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))

top_rated.plot(kind='bar', color='purple')

plt.title("Average Customer Rating by Brand")

plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```

```python
# 16. CHARGING SPEED ANALYSIS

plt.figure(figsize=(10,6))

sns.boxplot(
    x='market_segment',
    y='Charging Speed',
    data=df
)

plt.title("Charging Speed by Market Segment")

plt.show()
```

```python
# 17. HORSEPOWER DISTRIBUTION

plt.figure(figsize=(10,5))

sns.histplot(df['horsepower'], bins=20, color='green')

plt.title("Horsepower Distribution")

plt.show()
```

### STATISTICAL ANALYSIS

```python
# 18. CORRELATION MATRIX

corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# 19. SKEWNESS

print("\nSkewness")
print(df.skew(numeric_only=True))
```

```python
# 20. VARIANCE

print("Variance")
print(df.var(numeric_only=True))
```

```python
# 21. STANDARD DEVIATION

print("\nStandard Deviation")
print(df.std(numeric_only=True))
```

```python
# 22. PAIRPLOT

sns.pairplot(
    df[[
        'Price',
        'Battery Capacity',
        'range_miles',
        'customer_rating'
    ]]
)

plt.show()
```
